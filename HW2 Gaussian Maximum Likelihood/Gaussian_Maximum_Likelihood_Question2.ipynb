## Apply Batch Gradient Descent to compute linear regression

For linear regression, we need to compute intercept $θ_0$ and slope $θ_1$, so we have the prediction line $y_{predict}=θ_1 x + θ_0$

We have the lost function $$J(θ)=\dfrac{1}{2m}\times \sum_{n=1}^{100}\left(y_{predict}-y \right)^2$$

We update intercept and slope by repeating computing derivative of the lost function,
$θ := θ - η \times \dfrac{δJ(θ)}{δθ}$

In the meanwhile, we plot the lines.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x = np.array([8, 16, 22, 33, 50, 51])
y = np.array([5, 20, 14, 32, 42, 58])
m = len(x)

eta = 0.0001 # learning rate
iterations = 1000 # number of learning times

intercept_slope = [0, 0] # initiate intercept and slope

# Batch gradient descent
for i in range(iterations):
    # linear regression, hypothesis function
    y_predict = intercept_slope[0] + intercept_slope[1] * x

    if i < 30: # plot 30 learning lines
        color = mpl.colors.rgb2hex(plt.cm.BuGn(i / 40 + 0.1)) # color
        plt.plot(x, y_predict, linestyle="solid", color=color) # plot

    error = y_predict - y # error

    # derivative from the lost function
    gradient = [(1/m) * np.sum(error), (1/m) * np.sum(error * x)]

    # new learned intercept and slope
    intercept_slope[0] = intercept_slope[0] - eta * gradient[0]
    intercept_slope[1] = intercept_slope[1] - eta * gradient[1]

plt.xlabel('$X$', fontsize=16)
plt.ylabel('$Y$', fontsize=16, rotation=0)
plt.title('Relationship between $X$ and $Y$ by Batch Gradient Descent', fontsize=18)
plt.plot(x, y, 'o', color="blue")
plt.plot(x, y_predict, color="red", linewidth=2.5) # fitting line
plt.show()

print('Fitting line: Y = ', intercept_slope[1], 'X ', intercept_slope[0])
```
